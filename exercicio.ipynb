11 - O Registrador de Endereços da Memória (REM) é aquele capaz de determinar sua capacidade total, visto que 

2^tamanho do REM  *  Capacidade da Célula = Capacidade total

12 -Primeiramente, como tem 2^16 células, 16 será o tamanho do REM, que deve ser igual ao do CI.

Então, CI = 16 bits.

Como existem 256 instruções de máquina, o campo dos C. de Op. = 8 bits, sendo o tamanho da palavra = 24 bits (8 + 16 = 24)

Como a palavra = 24 bits, tem-se que:

RDM e Reg. = 24 bits.

Por fim, a memória tem 2^16 células de memória, e cada célula armazena 24 bits:

```python
total_capacidade = 2**16 * 24

print(f"A memória terá capacidade de {total_capacidade} bits, entre 2^20 e 2^21")
```
